# Modelización No-Supervisada : KMeans

# 1. Importar paquetes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

%config IPCompleter.greedy = True

pd.options.display.float_format = '{:.2f}'.format

import warnings
warnings.filterwarnings('ignore')
```

```python
nombre_df = 'df_tablon.pickle'
```

```python
df = pd.read_pickle(ruta + '/02_Datos/03_Trabajo/' + nombre_df)
```

# 3. Modelización

Primero creamos el espacio de las soluciones.

```python
min_k = 3
max_k = 7
soluciones = list(range(min_k, max_k))
```

Luego las listas donde se van a guardar los resultados, para luego analizarlos.

```python
codo = []
silueta = []
calins = []
davies = []
```

Ahora se realiza el analisis de las soluciones.

```python
for solucion in soluciones:
    
    #Instanciar
    cluster = KMeans(n_clusters = solucion, n_init=10)
    
    #Entrenar
    cluster.fit(df)
    
    #Métricas
    codo.append(cluster.inertia_)
    silueta.append(silhouette_score(df, cluster.labels_))
    calins.append(calinski_harabasz_score(df, cluster.labels_))
    davies.append(davies_bouldin_score(df, cluster.labels_))
```

Identificar la mejor solución.

```python
metricas = pd.DataFrame({'Solucion':soluciones,
              'Codo': codo,
              'Silueta':silueta,
              'Calins':calins,           
              'Davies':davies}).set_index('Solucion')

metricas
```

Ahora visualizar el rendimiento de las soluciones con las metricas.

```python
metricas.plot(subplots = True, figsize = (12,8), layout = (2,2), sharex = False)
plt.show()
```

**Analisis de las metricas:** De acuerdo con el primer grafico el "codo" se forma en los valores 4 y 5, pero en el valor 5 no muestra una mejora significativa. En el caso de la "silueta" los mejores valores estan ordenados con 3, 5 y 4, aunque la distancia entre ellos es pequeña. Por su parte, la metrica Calinski-Harabasz da mayor valor a la solucion que entrega clusters más separados y definidos, en nuestro caso esa solución seria el valor 3 seguido del 4. Por ultimo, Davies-Bouldi que entrega el valor más bajo a la solución con mayor distancia entre los clusters, dado el valor de 5 y 6 en esta ocasión.

**Decisión:** Lo ideal seria que las metricas convergan a la misma solución, pero no ha sido el caso. Por lo tanto, decidi tomar el valor de 4 clusters. El motivo es las 2 de las 4 metricas posicional dicho valor como el segundo mejor, mientras que las restantas lo posicionan como el tercero. Tambien es importante mencionar que se hicieron las pruebas con los otros valores y al final opte por el 4 dado los resultados de ellas.

Segmentación de la solución final.

```python
mejor_solucion = 4

#Instanciar
cluster = KMeans(n_clusters = mejor_solucion, n_init=10)
    
#Entrenar
cluster.fit(df)
```

Calculo de segmento y validación.

```python
df['segmento'] = cluster.predict(df)
```
